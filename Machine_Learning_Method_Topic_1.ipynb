```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define a function f(x) = x^2
def f(x):
    return x**2

# 2. Define the analytical gradient f'(x) = 2x
def gradient_f(x):
    return 2 * x

# 3. Visualization
x = np.linspace(-10, 10, 100)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='f(x) = x^2')

# Plot gradient vectors at specific points
for point in [-5, 0, 5]:
    grad = gradient_f(point)
    plt.arrow(point, f(point), 1, grad,
              head_width=0.5, head_length=1, fc='red', ec='red')
    plt.text(point, f(point) + 10, f"Grad={grad}", color='red')

plt.title("Visualizing the Gradient (Steepest Ascent)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# 1. Create dummy training data (1D points for simplicity)
# Format: [[feature1], [feature2], ...]
X_train = np.array([[1], [2], [10], [11]])
# Labels: Class 0 (Small numbers), Class 1 (Large numbers)
y_train = np.array([0, 0, 1, 1])

# 2. Initialize the k-NN model (k=3)
neigh = KNeighborsClassifier(n_neighbors=3)

# 3. "Train" the model (This just memorizes X_train and y_train)
neigh.fit(X_train, y_train)

# 4. Predict on a new point
new_point = np.array([[1.5]])
prediction = neigh.predict(new_point)

print(f"Input Point: {new_point[0][0]}")
print(f"Predicted Class: {prediction[0]}")
# Expected: Class 0, because 1.5 is closer to 1 and 2 than to 10 and 11.
```
